## Post 1500CE data

```python
drv="/Users/bizcocho/Desktop"
growperiod=drv+"/Geographical_Index/DataOthers/FAO/GAEZ/Agroecological/growingperiod/res01_lgd_crav6190.tif"
growperiod=gr.from_file(growperiod)
```

```python
energydir=drv+"/Geographical_Index/DataOthers/FAO/GAEZ/Agroclimatic/energyyield/rain_fed/"
tiffiles=[tiffile for tiffile in os.listdir(energydir) if tiffile.endswith('lo.tif')]
emat=np.zeros((len(tiffiles),growperiod.shape[0],growperiod.shape[1]))
```

```python
count=0
for tiffile in tiffiles:
    emat[count,:,:]=gr.from_file(energydir+tiffile).raster.data
    count+=1
emat=np.ma.masked_array(emat,mask=(emat==-9), fill_value=-9.)
```

```python
esum=emat.sum(axis=0)
emean=emat.mean(axis=0)
```

```python
esum= gr.GeoRaster(esum, growperiod.geot, projection=growperiod.projection, nodata_value=-9)
emean= gr.GeoRaster(emean, growperiod.geot, projection=growperiod.projection, nodata_value=-9)
esum.to_tiff('./data/post1500TotalCalories')
emean.to_tiff('./data/post1500AverageCalories')
post1500mean = gr.from_file('./data/post1500AverageCalories.tif')
post1500tot = gr.from_file('./data/post1500TotalCalories.tif')
```
